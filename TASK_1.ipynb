**Import only the libraries we need**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

**Load the Titanic data from internet Python**

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

print("First 5 rows:")
print(df.head())

print("\nMissing values right now:")
print(df.isnull().sum())
```

**Keep only the most important columns**

```python
df = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Survived']]

print("Columns we are keeping:")
print(df.columns.tolist())
```

**Fix missing values**

```python
# Age: fill missing with the middle value
df['Age'] = df['Age'].fillna(df['Age'].median())

# Embarked: fill the 2 missing with 'S' (most common port)
df['Embarked'] = df['Embarked'].fillna('S')

# Fare: very few missing → fill with 0 just in case
df['Fare'] = df['Fare'].fillna(0)

print("\nMissing values after fixing:")
print(df.isnull().sum())
```

**Change words (categories) to numbers**

```python
# Sex: male → 0, female → 1
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Embarked: S → 0, C → 1, Q → 2
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

print("\nAfter changing words to numbers:")
print(df.head())
```

**Scale (make numbers similar size) – Age and Fare only**

```python
scaler = StandardScaler()

df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\nAfter scaling (Age and Fare now look small):")
print(df[['Age', 'Fare']].head())
```

**Separate features (X) and target (y)**

```python
X = df.drop('Survived', axis=1)   # all columns except Survived
y = df['Survived']                # only Survived column

print("X shape:", X.shape)
print("y shape:", y.shape)
```

**Split into train and test parts**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 20% test, 80% train
    random_state=42      # same result every time
)

print("Training data shape:", X_train.shape)
print("Testing data shape :", X_test.shape)
```


**See if everything looks okay**


```python
print("First 5 rows of training data (X_train):")
print(X_train.head())

print("\nFirst 5 answers (y_train):")
print(y_train.head())
```

**try a very simple model**

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

score = model.score(X_test, y_test)
print("Model accuracy on test data:", round(score * 100, 1), "%")
```
